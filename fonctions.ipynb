# Fonctions et structures de contrôle

### 1 - Fonctions

```python
# Syntaxe: def nomDeLaFonction(arguments):
def celsiusToFarenheit(tempC):
  tempF = 1.8 * tempC + 32
  return tempF
```

```python
# Appel de la fonction
celsiusToFarenheit(30)
```

### 2 - Interactions console

```python
# input permet de récupérer une chaine entrée au clavier
tempC = int(input('Température en °C : '))
tempF = celsiusToFarenheit(tempC)
# print permet d'afficher des chaines et des variables
print('Température en °F :', tempF)
```

### 3 - Conditions

```python
# Syntaxe: if condition:
valeur = int(input('Entrez un entier:'))
if valeur%2 == 0:
  print(valeur, ' est un nombre pair')
else:
  print(valeur, ' est un nombre impair')

# La tabulation permet de délimiter un bloc de code (équivalent des accolades)
```

### 4 - Boucles

```python
# Avec un nombre prédéfini ou calculé
for i in range(9):
  print('2 puissance', i, '=', 2**i)

# La fonction range(n) renvoie un itérateur de 0 à n-1
```

```python
# Avec une condition
while i != -1:
  i = int(input('Entrez un entier:'))
  print('2 puissance', i, '=', 2**i)
```
